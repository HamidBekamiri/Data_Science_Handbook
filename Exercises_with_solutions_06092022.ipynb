```python
# 1. df.iloc() & df.loc()

df_review_id_price = df_review.loc[:, ['course_id', 'price']]
df_review_id_subject = df_review.loc[:, ['course_id', 'subject', 'num_reviews']]
```

```python
# 3. df.dropna()

df_review_id_subject.dropna(inplace=True)
```

```python
# 4. df.fillna()

df_review_id_subject.fillna(0)
```

```python
df_review_id_subject.isnull().sum()
df_review_id_subject.fillna(0, inplace=True)
df_review_id_subject.isnull().sum()
```

```python
# 5. df.merge()

df_subject_price = pd.merge(df_review_id_subject, df_review_id_price, on='course_id', how='inner')
```

```python
df_subject_price.subject.value_counts()
```

```python
# 6. df.concat()

df_subject_price_part1 = df_subject_price.iloc[:1000, :]
df_subject_price_part2 = df_subject_price.iloc[1000:, :]
```

```python
pd.concat([df_subject_price_part1, df_subject_price_part2])
```
